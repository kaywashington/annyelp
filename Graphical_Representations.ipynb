```python
import pandas as pd

nyt_logreg = {
    "precision": [0.75, 0.0, 0.38],
    "recall": [0.3, 0.0, 1.0],
    "f1-score": [0.43, 0.0, 0.56]
}

nyt_lstm = {
    "precision": [0.5, 0.67, 0.31],
    "recall": [0.7, 0.2, 0.4],
    "f1-score": [0.58, 0.31, 0.35]
}

yelp_logreg = {
    "precision": [0.81, 0.51, 0.79],
    "recall": [0.53, 0.15, 0.98],
    "f1-score": [0.64, 0.23, 0.88]
}

yelp_lstm = {
    "precision": [0.53, 0.26, 0.87],
    "recall": [0.59, 0.41, 0.31],
    "f1-score": [0.87, 0.73, 0.8]
}
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
labels = ['Negative', 'Neutral', 'Positive']
x = np.arange(len(labels))  # [0, 1, 2]
width = 0.35


# Plotting
fig, axes = plt.subplots(3, 1, figsize=(8, 15))

# Precision Bar Chart
axes[0].bar(x - width/2, yelp_logreg["precision"], width, label='LogReg')
axes[0].bar(x + width/2, yelp_lstm["precision"], width, label='LSTM')
axes[0].set_xlabel('Class')
axes[0].set_ylabel('Precision')
axes[0].set_title('Yelp Precision: LogReg vs LSTM')
axes[0].set_xticks(x)
axes[0].set_xticklabels(labels)
axes[0].set_ylim(0, 1)
axes[0].legend()

# Accuracy Bar Chart
axes[1].bar(x - width/2, yelp_logreg["recall"], width, label='LogReg')
axes[1].bar(x + width/2, yelp_lstm["recall"], width, label='LSTM')
axes[1].set_xlabel('Class')
axes[1].set_ylabel('Accuracy')
axes[1].set_title('Yelp Accuracy: LogReg vs LSTM')
axes[1].set_xticks(x)
axes[1].set_xticklabels(labels)
axes[1].set_ylim(0, 1)
axes[1].legend()

axes[2].bar(x - width/2, yelp_logreg["f1-score"], width, label='LogReg')
axes[2].bar(x + width/2, yelp_lstm["f1-score"], width, label='LSTM')
axes[2].set_xlabel('Class')
axes[2].set_ylabel('F1-Score')
axes[2].set_title('Yelp F1-Score: LogReg vs LSTM')
axes[2].set_xticks(x)
axes[2].set_xticklabels(labels)
axes[2].set_ylim(0, 1)
axes[2].legend()

# Adjust layout
plt.tight_layout()
plt.show()
```
